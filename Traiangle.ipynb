```python
from dot import Dot
import random
import matplotlib.pyplot as plt
```

```python
def random_point_in_triangle(A, B, C):
    # Generate random barycentric coordinates
    u, v = random.random(), random.random()
    if u + v > 1:  # Adjust to ensure the point is inside the triangle
        u, v = 1 - u, 1 - v
    
    # Calculate the random point
    x = (1 - u - v) * A.x + u * B.x + v * C.x
    y = (1 - u - v) * A.y + u * B.y + v * C.y
    return Dot(x, y)

def plot_dots(dots, triangle_vertices):
    # Extract x and y coordinates for all dots
    x_coords = [dot.x for dot in dots]
    y_coords = [dot.y for dot in dots]

    # Extract triangle vertices for plotting
    tri_x = [vertex.x for vertex in triangle_vertices] + [triangle_vertices[0].x]
    tri_y = [vertex.y for vertex in triangle_vertices] + [triangle_vertices[0].y]

    # Plot the points and triangle
    plt.figure(figsize=(20, 20))
    plt.scatter(x_coords, y_coords, s=1, c='blue', label='Random Points')  # Small blue dots
    plt.plot(tri_x, tri_y, c='red', label='Triangle')  # Red triangle edges
    plt.title("Random Points Inside a Triangle")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.legend()
    plt.grid(True)
    plt.axis('equal')  # Maintain aspect ratio
    plt.show()
```

```python
A = Dot(0, 0)
B = Dot(0, 10)
C = Dot(10, 5)

corners = [A, B, C]

points = corners.copy()
```

```python
plot_dots(points, corners)
```

```python
current_point = random_point_in_triangle(A, B, C)
for i in range(1000):
    current_point = (current_point + random.choice(corners)) * 0.5
    points.append(current_point)
```

```python
plot_dots(points, corners)
```
